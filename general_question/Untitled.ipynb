```python
# data pre processing

revenue_mapping = {
    "$1M - $5M": "SMEs",
    "$5M - $10M": "SMEs",
    "$10M - $50M": "SMEs",
    "$50M - $100M": "Medium Enterprises",
    "$100M - $500M": "Large Enterprises",
    "$250M - $500M": "Large Enterprises",
    "$1B+": "Large Enterprises",
}

df_ai["Revenue"] = df_ai["Revenue"].map(revenue_mapping)
```

```python
# Define a function to extract numerical values
def extract_funding_amount(text):
    # Use regular expressions to find the numerical part
    match = re.search(r'\$([\d.,]+)[MB]', text)
    if match:
        funding = match.group(1)
        # Remove commas and convert to float
        return float(funding.replace(',', ''))
    else:
        return None

# Apply the function to create a new column "funding_amount"
df_ai["Most Recent Funding Amount"] = df_ai["Most Recent Funding Amount"].apply(extract_funding_amount)
```

```python
# Define a function to convert funding amounts to numerical values
def convert_funding_amount(text):
    # Use regular expressions to find the numerical part
    match = re.search(r'\$([\d.,]+)\s*(million|thousand)?', text, re.I)
    if match:
        amount = match.group(1)
        unit = match.group(2)

        if unit and unit.lower() == "million":
            return float(amount) * 1e6
        elif unit and unit.lower() == "thousand":
            return float(amount) * 1e3
        else:
            return float(amount)
    else:
        return None

# Apply the function to create a new column "numerical_funding"
df_ev["Revenue"] = df_ev["Revenue"].apply(convert_funding_amount)
```
